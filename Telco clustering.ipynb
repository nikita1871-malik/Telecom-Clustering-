```python
from sklearn.preprocessing import normalize
data_scaled = normalize(telco)
data_scaled = pd.DataFrame(data_scaled, columns=telco.columns)
data_scaled.head() 
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))  
plt.title("Dendrograms") 
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
```

```python
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))

plt.axhline(y=5.3, color='r', linestyle='--')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
```

```python
cl=cluster.fit_predict(data_scaled)
```

```python
clus3=telco.groupby(["Cluster"]).agg({"region":["mean"],
                                "tenure":["mean"],
                                "age":["mean"],
                                "marital":["mean"],
                                "address":["mean"],
                                "income":["mean"],
                                     "ed":["mean"],
                                     "employ":["mean"],
                                     "retire":["mean"],
                                     "gender":["mean"],
                                     "reside":["mean"],
                                     "tollfree":["mean"],
                                     "equip":["mean"],
                                     "callcard":["mean"],
                                     "wireless":["mean"],
                                     "multline":["mean"],
                                     "voice":["mean"],
                                     "pager":["mean"],
                                     "internet":["mean"],
                                     "callid":["mean"],
                                    "callwait":["mean"],
                                     "forward":["mean"],
                                     "confer":["mean"],
                                     "longmon":["mean"],
                                     "tollmon":["mean"], 
                                     "equipmon":["mean"],
                                     "cardmon":["mean"],
                                     "wiremon":["mean"],
                                     "ebill":["mean"],
                                     "custcat":["mean"],
                                     "Cluster":["mean"]
                                    })
new_col=[]
for i in range(len(list(clus3.columns))):
    new_col.append("_".join(list(clus3.columns[i])))
    
#     print("_".join(list(clus4.columns[i])))
clus3.columns=new_col
cluster3=clus3.T
cluster3.columns=["Clust0", "Clust1", "Clust2"]
```

```python
clus3.T
```

```python
cluster3_1=cluster3.reset_index()
cluster3_1.drop(columns=["index"], inplace=True)
cluster3_1
```

```python
d_1=telco[['region', 'tenure', 'age', 'marital', 'address', 'income', 'ed',
       'employ', 'retire', 'gender', 'reside', 'tollfree', 'equip', 'callcard',
       'wireless', 'multline', 'voice', 'pager', 'internet', 'callid',
       'callwait', 'forward', 'confer', 'longmon', 'tollmon', 'equipmon',
       'cardmon', 'wiremon', 'ebill', 'custcat', 'Cluster']].describe().T["mean"].reset_index()
```

```python
d_1
```

```python
clust=pd.concat([d_1, cluster3_1], axis=1)
```
